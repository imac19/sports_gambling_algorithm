## Algorithm One: Simple ELO rating comparison
For the first algorithm, simply going to compare fivethirtyeight ELO rating win probabilities to Vegas moneyline implied probabilities to see how we would fare only betting on games with a large gap between those two probabilities. 

```python
# Functions:
# get_prob_advantage returns two values, the best probability gap for the game, as well as the team given the advantage
# by the probability gap
# place_bet_100 returns True or False based on a given threshold: any team with a probability advantage higher than 
# threshold will have a 100 dollar bet placed on them

def get_prob_advantage(game):
    advantage_home = game['elo_prob1'] - game['implied_prob_home_open']
    advantage_away = game['elo_prob2'] - game['implied_prob_away_open']
    
    if advantage_home > advantage_away:
        return advantage_home, 'home'
    else:
        return advantage_away, 'away'
    
def place_bet_100(game, threshold):
    if game['best_prob_advantage'] >= threshold:
        return True
    else:
        return False
```

```python
# Apply above functions, test threshold for now of .03, will experiment with that value later

df = df.join(pd.DataFrame(df.apply(lambda x: get_prob_advantage(x), axis=1).tolist(), columns=['best_prob_advantage', 'advantage_team']))
```

```python
df
```

```python
# Function to get returns based on probability advantage and 100 dollar bet

def get_return_100(game):
    if game.advantage_team == 'home':
        return game['Home ML Bet Return']
    else:
        return game['Away ML Bet Return']
```

```python
# Check returns from above method

bets_100_03 = df.copy()
bets_100_03['place_bet_100'] = bets_100_03.apply(lambda x: place_bet_100(x, threshold=.03), axis=1)
bets_100_03 = bets_100_03[bets_100_03.place_bet_100 == True]
bets_100_03['returns'] = bets_100_03.apply(lambda x: get_return_100(x), axis=1)
bets_100_03.groupby(by=['season']).sum()['returns'].reset_index().plot.bar(x='season', y='returns')
```

## Algorithm Two: Simple QB Adjusted ELO rating comparison
For the second algorithm, simply going to compare fivethirtyeight QB adjsuted ELO rating win probabilities to Vegas moneyline implied probabilities to see how we would fare only betting on games with a large gap between those two probabilities. 


```python
# Functions:
# get_prob_advantage returns two values, the best probability gap for the game, as well as the team given the advantage
# by the probability gap
# place_bet_100 returns True or False based on a given threshold: any team with a probability advantage higher than 
# threshold will have a 100 dollar bet placed on them

def get_prob_advantage_qb(game):
    advantage_home = game['qbelo_prob1'] - game['implied_prob_home_open']
    advantage_away = game['qbelo_prob2'] - game['implied_prob_away_open']
    
    if advantage_home > advantage_away:
        return advantage_home, 'home'
    else:
        return advantage_away, 'away'
    
def place_bet_100_qb(game, threshold):
    if game['qb_best_prob_advantage'] >= threshold:
        return True
    else:
        return False
```

## QB Adjusted ELO Predictive Accuracy
How accurate is QB Adjusted ELO? What is the accuracy threshold we are looking to beat?

```python
# Functions to get qb elo predicted winner, check accuracy against actual results

def get_pick_qb_elo(game):
    if game.qbelo_prob1 >= game.qbelo_prob2:
        return 'Home'
    else:
        return 'Away'
    
def is_pick_correct_qb_elo(game):
    if game.pick_qb_elo == game['Winner HA']:
        return 1 
    else:
        return 0
    
def get_pick_basic_elo(game):
    if game.elo_prob1 >= game.elo_prob2:
        return 'Home'
    else:
        return 'Away'
    
def is_pick_correct_basic_elo(game):
    if game.pick_basic_elo == game['Winner HA']:
        return 1 
    else:
        return 0

df['pick_qb_elo'] = df.apply(lambda x: get_pick_qb_elo(x), axis=1)
df['is_pick_correct_qb_elo'] = df.apply(lambda x: is_pick_correct_qb_elo(x), axis=1)
df['pick_basic_elo'] = df.apply(lambda x: get_pick_basic_elo(x), axis=1)
df['is_pick_correct_basic_elo'] = df.apply(lambda x: is_pick_correct_basic_elo(x), axis=1)
df
```

```python
# Get season by season accuracy

df.groupby(by=['season']).mean()[['is_pick_correct_qb_elo', 'is_pick_correct_basic_elo']].reset_index()
```

```python
df[df.season != 2021].mean()[['is_pick_correct_qb_elo', 'is_pick_correct_basic_elo']]
```

```python
# Bar chart

df.groupby(by=['season']).mean()[['is_pick_correct_qb_elo', 'is_pick_correct_basic_elo']].reset_index().plot.bar(x='season', y=['is_pick_correct_qb_elo', 'is_pick_correct_basic_elo'])
```

```python
# Get total accuracy

df.mean()[['is_pick_correct_qb_elo', 'is_pick_correct_basic_elo']]
```

Conclusion: Per season accuracy seems to fluctuate between 60 and 70 percent, with an overall accuracy of 66%. Goal for the new algorithm is to be able to beat 66% accuracy, or at the very least to have predictions that are more confident but still as accurate.
