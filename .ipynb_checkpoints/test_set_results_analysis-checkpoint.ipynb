```python
# def get_ml_weighted_avg(game):
    
#     if game.advantage_team_algo == 'home':
#         return game.best_prob_advantage_algo * (game.home_win_prob * 2)
#     else:
#         return game.best_prob_advantage_algo * (game.away_win_prob * 2)

def get_return_proportional_ml(game):
    
    bet_weight = 0
    
    # Only use first one, game.best_prob_advantage_algo
    
    if game.best_prob_advantage_algo >= .2:
        bet_weight = 1

#     elif game.ml_weighted_avg >= .2:
#         bet_weight = 1
        
    if game.advantage_team_algo == 'home':
        return bet_weight * game['Home ML Bet Return'], bet_weight * 100
    else:
        return bet_weight * game['Away ML Bet Return'], bet_weight * 100
    
def get_return_proportional_ou(game):
    
    bet_weight = 0
    
    if abs(game.total_score_diff_algo) >= 3 :
        bet_weight = 1
        
    return bet_weight * game['{}_bet_return'.format(game.ou_advantage)], bet_weight*100


def get_return_proportional_spread(game):
    
    bet_weight = 0
    
    if game.best_spread_advantage_algo > 5:
        bet_weight = 1

#     if game.spread_advantage_team_algo == 'Home':
#         if game.implied_prob_home_open > .5:
#             bet_weight = 1
#     else:
#         if game.implied_prob_home_open < .5:
#             bet_weight = 1
            
    if game.spread_advantage_team_algo == 'Home':
        return bet_weight * game['Home Spread Bet Return'], bet_weight * 100
    else:
        return bet_weight * game['Away Spread Bet Return'], bet_weight * 100
    
gambling_df = combined_results.copy()
# gambling_df['ml_weighted_avg'] = gambling_df.apply(lambda x: get_ml_weighted_avg(x), axis=1)
gambling_df[['moneyline_returns', 'moneyline_wagered']] = gambling_df.apply(lambda x: get_return_proportional_ml(x), axis=1, result_type='expand')
gambling_df[['spread_returns', 'spread_wagered']] = gambling_df.apply(lambda x: get_return_proportional_spread(x), axis=1, result_type='expand')
gambling_df[['ou_returns', 'ou_wagered']] = gambling_df.apply(lambda x: get_return_proportional_ou(x), axis=1, result_type='expand')
gambling_df['total_returns'] = gambling_df.moneyline_returns + gambling_df.spread_returns + gambling_df.ou_returns
gambling_df['total_wagered'] = gambling_df.moneyline_wagered + gambling_df.spread_wagered + gambling_df.ou_wagered

gambling_df
```

Note: The moneyline and point spread algorithms can sometimes "disagree" with each other, i.e. the moneyline algorithm will predict the home team to win, while the point spread algorithm will predict that the away team should be favored. However, this doesn't seem to be much of an issue, and the bets don't necessarily counteract each other. As an example, take the titans 49ers game from 2021. The point spread algorithm identifies the titans should be favored by a point, when they are actually the underdog, thus indicating that the titans are more likely to win. However, the moneyline algorithm gives the titans only a 44% chance to win, which doesnt match the results of the point spread algorithm. However, the vegas implied odds have the titans at a 40% chance to win. So even though the predictions are conflicting, the actual suggested bets are not, as they have you taking both titans spread and titans moneyline. This scenario happens somewhat often, and doesnt seem to hinder the algorithm performance in any way. In most of these cases, only one of the algorithms is confident enough to place a bet. In most of the cases where both are confident, they usually match like the scenario above. In the rare case where they do not match up, it is not advised to take any of the bets.
